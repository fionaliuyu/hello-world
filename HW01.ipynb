## Problem 1 (30 points)

Given observations $\bX\;$ and  $\by$, and using equation (10) in lecture 3, we see that the fitted values $\bhy\,$ are given by
$$
\bhy = \bX \cdot ( \bX\;^T \cdot \bX\;)^{-1} \cdot \bX\;^T  \cdot \by
$$
We define the $n\times n$ matrix operator
$$
P := \bX \cdot ( \bX\;^T \cdot \bX\;)^{-1} \cdot \bX\;^T \qquad (1)
$$
and its complement
$$
M := \bid - P \qquad (2)
$$
with $\bid\,$ the $n \times n$ identity matrix.  
The matrix $P$ is an **orthogonal projection** matrix. It projects any vector $\by\,$ into $P\cdot \by$, which lives in the subspace spanned by the column vectors of $\bX$.  

Show the above statement, by proving the following:  
1) $P^2=P$ (projecting twice is the same as projecting once, idempotent matrix)  
2) $(P\cdot\bx)^T \cdot (M \cdot \by)= 0$ (the projections of $P$ and $M$ are orthogonal vectors for every $\bx$ and $\by$)

Therefore, we have the linear regression model  
$$
r_{stk} - r_{rf} = \alpha + \beta (r_{mkt} - r_{rf}) + \epsilon \qquad (3)
$$

```python
import pandas as pd
import numpy as np
from sklearn import  linear_model
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
datapath = '/Users/yuliu/Documents/7773 Machine Learning/Data/stock-treasury-2004_2006.csv'
df = pd.read_csv(datapath)
df_stock_price = pd.read_csv(datapath)
daily_t_return = df['TREAS_3M']/90
#two list to store returns
stock_returns = []
totRows = df_stock_price.shape[0]
df_stock_price.drop(columns = ['Date','TREAS_3M'],inplace = True) #take only stocks prices and SP
```

```python
#calculate the return, include Stocks and SP
for k in range(totRows-1):
    stock_returns.append(((df_stock_price.iloc[k+1]/df_stock_price.iloc[k])-1.0).values.tolist())
```

```python
# trun return into df
df_stock_return = pd.DataFrame(np.array(stock_returns).reshape(672,11), columns = df_stock_price.columns)
df_stock_return=df_stock_return*10000
```

```python
#turn the bond returns into bp scale(already divided by 90), then drop the first return, to match the size of stock returns
daily_t_return = daily_t_return*100
daily_t_return.drop(0,inplace = True)
```

```python
#substrct the daily Treasury yield from stock returns and sp returns 
daily_t_return.index = range(len(daily_t_return))
df_stock_net_return = df_stock_return.sub(daily_t_return,axis='index')
df_sp_net_return = df_stock_net_return['SP']
df_stock_net_return.drop(columns = 'SP', inplace = True)
```

```python
# use sk-learn model 
y=df_stock_net_return
x=df_sp_net_return
skl_result = []
sta_result = []
lm = linear_model.LinearRegression(fit_intercept=True)
for i in range(y.shape[1]):
    lm.fit(x[:,np.newaxis],y.iloc[:,i] )   # NOTE: fitting modifies the object
    r_sq = lm.score(x[:,np.newaxis],y.iloc[:,i])
    alpha = lm.intercept_
    beta = np.asscalar(lm.coef_)
    skl_result.append([alpha,beta,r_sq])
skl = pd.DataFrame(skl_result,index =df_stock_net_return.columns,columns = ['alpha','beta','r_squared'])
skl.reset_index(inplace=True)
skl.columns=['sym','alpha','beta','r_squared']
skl
```

```python
#use statsmodels
X = sm.add_constant(x)
for i in range(y.shape[1]):
    mdl =sm.OLS(y.iloc[:,i],X)  
    res = mdl.fit()
    alpha = res.params[0]
    beta = res.params[1]
    alpha_se = res.bse[0] 
    beta_se = res.bse[1]
    alpha_pval = res.pvalues[0]
    beta_pval= res.pvalues[1]
    sta_result.append([alpha,alpha_se,alpha_pval,beta,beta_se,beta_pval,res.rsquared])
sta = pd.DataFrame(sta_result,index =df_stock_net_return.columns,columns = ['alpha','alpha_se','alpha_pval','beta','beta_se','beta_pval','r_squared'])
sta.reset_index(inplace=True)
sta.columns=['sym','alpha','alpha_se','alpha_pval','beta','beta_se','beta_pval','r_squared']
sta
```

Compare the two data frames and assert that they return the same values for alpha and beta.

```python
# compare two results
column = ['alpha','beta']
compare = (skl[column].round(4) == sta[column].round(4))
compare
```

Based on the p-values, can we reject the null hypotheses alpha=0 or beta=0? 

Yes, at the confidence level of 95%, we can reject the null hypothesis.
